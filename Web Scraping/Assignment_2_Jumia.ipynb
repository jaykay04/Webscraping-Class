```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
import pandas as pd
```

```python
url = "https://www.jumia.com.ng/catalog/?q=shoes+for+men"
ua = UserAgent()
userAgent = ua.random
headers = {"user-agent": userAgent}
page = requests.get(url, headers = headers)

page
```

```python
soup = BeautifulSoup(page.content, "html.parser")
soup.find("div", class_="info")
len(soup.text)
```

```python
shoe_collection = soup.find_all("div", class_ = "info")
```

```python
shoe_types = []
prices = []
ratings = []


for shoe in shoe_collection:
    #shoe_types
    if shoe.find("h3", class_ = "name") is not None:
        shoe_type = shoe.find("h3", class_ = "name")
        shoe_types.append(shoe_type.text)
    else:
        shoe_types.append("None")
        
    #prices
    if shoe.find("div", class_ = "prc") is not None:
        price = shoe.find("div", class_ = "prc")
        prices.append(price.text)
    else:
        prices.append("None")
        
    #ratings
    if shoe.find("div", class_ = "rev") is not None:
        rating = shoe.find("div", class_ = "rev")
        ratings.append(rating.text.replace(" out of 5(", "-"))
    else:
        ratings.append("None")
        
```

```python
shoe_collection_df = pd.DataFrame(
    {
        "shoe_type": shoe_types,
        "price": prices,
        "rating": ratings
    }
)

shoe_collection_df
```

```python
shoe_collection_df["rating"] = shoe_collection_df["rating"].astype(str)
```

```python
shoe_collection_df.dtypes
```

```python
shoe_collection_df = shoe_collection_df[shoe_collection_df["rating"] != "None"]
shoe_collection_df
```

```python
shoe_collection_df[["rating_value", "rating_count"]] = shoe_collection_df["rating"].str.split("-", expand = True)
shoe_collection_df
```

```python
shoe_collection_df["rating_count"] = shoe_collection_df["rating_count"].str.replace(")", "")
shoe_collection_df
```

```python
shoe_collection_df.drop("rating", axis = 1, inplace = True)
shoe_collection_df
```

```python
shoe_collection_df.to_csv("shoe_collection_data.csv", index = False)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.bbc.com/sport/football/premier-league/table"
page = requests.get(url)
soup = BeautifulSoup(page.content, "html.parser")
```

```python
epl_table = soup.find("table", class_ = "ssrcss-5o3au1-Table e3bga5w2")
```

```python
headers = []


for i in epl_table.find_all("th"):
    header = i.text
    headers.append(header)
    
epl_df = pd.DataFrame(columns = headers)
epl_df

for j in epl_table.find_all("tr")[1:]:
    record = j.find_all("td")
    record_row = [x.text for x in record]
    length = len(epl_df)
    epl_df.loc[length] = record_row
    
epl_df.drop(["Form, Last 6 games, Oldest first"], axis = 1, inplace =True)
epl_df
```

```python
epl_df["Position"] = epl_df["Team"].str.extract(r"^(\d+)")
epl_df["Team"] = epl_df["Team"].str.extract(r"^\d+(.*)")
epl_df
```

```python
cols = ['Position'] + [col for col in epl_df.columns if col != 'Position']
epl_df = epl_df[cols]
epl_df
```

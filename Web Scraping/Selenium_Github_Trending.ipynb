```python
from selenium.webdriver import Chrome
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support import expected_conditions as EC
from fake_useragent import UserAgent
#from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
from selenium.webdriver.chrome.service import Service
import time
import requests  # For making HTTP requests
from bs4 import BeautifulSoup  # For parsing HTML content
from fake_useragent import UserAgent  # For generating random user agents
import pandas as pd  # For data manipulation and creating DataFrames
import numpy as np
import concurrent.futures
from selenium.common.exceptions import NoSuchElementException
```

```python
service = Service(executable_path = r'C:\Users\LENOVO\Downloads\chromedriver-win64\chromedriver-win64\chromedriver.exe')
driver = webdriver.Chrome(service = service)

driver.get("https://github.com/trending")

try:
    date_dropdown = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//details[@id = 'select-menu-date']")))
    date_dropdown.click()
    time.sleep(3)
    
    this_month_option = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//span[contains(text(), 'This month')]")))
    this_month_option.click()
    time.sleep(5)
    
    language_dropdown = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//details[@id = 'select-menu-language']")))
    language_dropdown.click()
    time.sleep(3)
    
    python_option = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//span[contains(text(), 'Python')]")))
    python_option.click()
    time.sleep(5)
except Exception as e:
    print(e)

# Find the links to the data we want
repos = driver.find_elements(By.XPATH, "//h2[@class = 'h3 lh-condensed']") # XPATH can take spaces

descriptions = driver.find_elements(By.CLASS_NAME, 'col-9.color-fg-muted.my-1.pr-4')
# Using CLASS_NAME, you only have to input the class name and remove spaces, change the spaces to dot

stars = driver.find_elements(By.XPATH, ".//a[contains(@href, '/stargazers')]")

forks = driver.find_elements(By.XPATH, ".//a[contains(@href, '/forks')]")

# Next, we are getting the text contents
repo_names = [repo.text.strip() for repo in repos]
repo_descriptions = [desc.text.strip() for desc in descriptions]
repo_stars = [star.text.strip() for star in stars]
repo_forks = [fork.text.strip() for fork in forks]

#Lets ceck the lenght of the Lists
print(f"Repository Names: {len(repo_names)}")
print(f"Descriptions: {len(repo_descriptions)}")
print(f"Stars: {len(repo_stars)}")
print(f"Forks: {len(repo_forks)}")

#Lets pad shorter list with None for consistency
max_length = max(len(repo_names), len(repo_descriptions), len(repo_stars), len(repo_forks))

# Adjust lengths
repo_names += ["None"] * (max_length - len(repo_names))
repo_descriptions += ["None"] * (max_length - len(repo_descriptions))
repo_stars += ["None"] * (max_length - len(repo_stars))
repo_forks += ["None"] * (max_length - len(repo_forks))

github_df = pd.DataFrame(
    {
        "repository": repo_names,
        "descriptions": repo_descriptions,
        "stars": repo_stars,
        "forks": repo_forks
    }
)

driver.quit()
github_df
```

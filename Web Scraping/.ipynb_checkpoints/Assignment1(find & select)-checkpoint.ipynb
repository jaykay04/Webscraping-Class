```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/Qatar"

response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
```

### find & find_all

```python
title = soup.find("title")
title.text
```

```python
headers2 = soup.find_all("div", class_ = "mw-heading mw-heading2" )

for subtopics in headers2:
    print(subtopics.text)
```

```python
headers3 = soup.find_all("div", class_ = "mw-heading mw-heading3")
for subtopics in headers3:
    print(subtopics.text)
```

## select_one & select

```python
body = soup.select_one("div")
body.text
```

```python
body1 = soup.select("div")

for x in body1:
    print(x.text)
```

```python
test1 = soup.select_one("div.mw-heading")
test1.text
```
